```python
# Dependencies
import pandas as pd
import scipy.stats as st
import numpy as np
import json
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
us_like_path = "output/cleaned_us_data.csv"
```

```python
us_like_df = pd.read_csv(us_like_path)
us_like_df.info()
```

```python
len(us_like_df['video_id'].unique())
```

```python
# Investigating the effect of the number of the views on the number of the likes for entertainment category
us_like_24 = us_like_df.loc[us_like_df['category_id'] == 24]
us_like_24_one_trend = us_like_24.loc[us_like_24.duplicated(keep='first')]
us_like_24_one_trend.head()
x_values = us_like_24_one_trend["views"]/1000000
y_values = us_like_24_one_trend["likes"]/1000
plt.scatter(x_values,y_values)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
r_value = "R = " + str(round(rvalue, 3))
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(1,350),fontsize=15,color="red")
plt.annotate(r_value,(1.5,300),fontsize=15,color="red")
plt.xlabel('Views of the Video (in millions)')
plt.ylabel('Likes for the Video (in thousands)')
plt.title('Category of Entertainment')
plt.show()
```
